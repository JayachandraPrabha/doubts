```python
def cal_vif(X):
    from statsmodels.stats.outliers_influence import variance_inflation_factor
    vif=pd.DataFrame()
    vif['Variables']=X.columns
    vif['VIF']=[variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
    return vif
# Minimal multicolinearity is requried foor a good model
# VIF value is =1, no correlation, VIF<3 to 5 less correlation, VIF>5 more correlated values
# created a function & a table whose columns as variables & VIF, 
# with the help of the varaiance inflation factor, considering the values in the columns of the dataset, we can put VIF values
```

# T-test

Independent samples --> Unpaired T-Test

Group= Different

Condition= Same

Dependent samples --> Paired T-Test

Group= Same

Condition= Different

```python
dataset
```

```python
dataset.dropna()
```

```python
# dataset[dataset['gender']=='M']
dataset[dataset['gender']=='M']['salary']
```

```python
# Independent - Unpaired T-Test (Group-same,condition-different)
# Group-Gender(M/F), Condition-Salary

from scipy.stats import ttest_ind
male=dataset[dataset['gender']=='M']['salary']
female=dataset[dataset['gender']=='M']['salary']
ttest_ind(male,female)
```

```python
from scipy.stats import ttest_ind
central=dataset[dataset['ssc_b']=='Central']['salary']
others=dataset[dataset['ssc_b']=='Others']['salary']
ttest_ind(central,others)
```

```python
# dependent - paired t-test (Group==different, Condition=same)
# Group=M, condition=ssc_p,hsc_p

from scipy.stats import ttest_rel
male1=dataset[dataset['gender']=='M']['ssc_p']
male2=dataset[dataset['gender']=='M']['hsc_p']
ttest_rel(male1,male2)
```

```python
from scipy.stats import ttest_rel
com1=dataset[dataset['hsc_s']=='Commerce']['etest_p']
com2=dataset[dataset['hsc_s']=='Commerce']['mba_p']
ttest_rel(com1,com2)
```

# Hypothesis testing

```python
# Hypothesis testing: Null hypothesis:(-), alternate hypothesis:(+)
# if p-value<0.05 --> reject the null hypothesis, accept the alternate hypothesis
```

```python
# Statement: SSC_P & HSC_P of male have similarity?
# Null hypothesis: SSC_P & HSC_P of male have no similarity
# Alternate hypothesis: SSC_P & HSC_P of male have similarity
```

```python
# dependent--> paired ttest --> group-different, condition-same, n-c
from scipy.stats import ttest_rel
male1=dataset[dataset['gender']=='M']['ssc_p']
male2=dataset[dataset['gender']=='M']['hsc_p']
ttest_rel(male1,male2)
```

```python
# pvalue: 0.49>0.05, condition fails so null hypothesis is rejected & alternate hypothesis is selected
# so, ssc_p & hsc_p of male have similarity
```
