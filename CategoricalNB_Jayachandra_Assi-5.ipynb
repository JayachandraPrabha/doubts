```python
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
dataset.shape
```

```python
dataset.value_counts()
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
independent
```

```python
dependent=dataset[['classification_yes']]
dependent
```

```python
dataset['classification_yes'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# from sklearn.preprocessing import StandardScaler
# sc=StandardScaler()
# x_train=sc.fit_transform(x_train)
# y_train=sc.fit(y_train)
```

```python
from sklearn.naive_bayes import CategoricalNB
classifier=CategoricalNB()
classifier.fit(x_train,y_train)
# from sklearn.model_selection import GridSearchCV
# grid_params={'penalty':['l1', 'l2', 'elasticnet', None],
#             'solver':['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
#             'multi_class':['auto', 'ovr', 'multinomial']}
# grid=GridSearchCV(classifier,grid_params,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
# grid.fit(x_train,y_train)
# penalty{‘l1’, ‘l2’, ‘elasticnet’, None}
# solver{‘lbfgs’, ‘liblinear’, ‘newton-cg’, ‘newton-cholesky’, ‘sag’, ‘saga’}
# multi_class{‘auto’, ‘ovr’, ‘multinomial’}
```

```python
# result=grid.cv_results_
# result
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
# solution=grid.best_params_
# solution
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test, y_pred)
print(clf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_score=roc_auc_score(y_test, classifier.predict_proba(x_test)[:,1])
roc_score
```
